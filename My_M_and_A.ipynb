```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
data1 = pd.read_csv("/content/Data/only_wood_customer_us_1.csv")
data2 = pd.read_csv("/content/Data/only_wood_customer_us_2.csv", sep=";", header=None, names=['Age','City','Gender','FirstName LastName','Email'])
data3 = pd.read_table("/content/Data/only_wood_customer_us_3.csv",sep="\t|,")
```

## **Data1**

```python
data1 = data1.drop(["UserName"], axis = 1)
data1['Country'] = 'USA'
```

```python
data1['FirstName'] = data1['FirstName'].str.title()

data1['LastName'] = data1['LastName'].str.title()

data1['City'] = data1['City'].str.capitalize()

data1['Email'] = data1['Email'].str.lower()
```

```python
gender = {"0": "Female", "1" : "Male", "F" : "Female", "M" : "Male"}

data1['Gender'] = data1['Gender'].replace(gender)

data1['FirstName'] = data1['FirstName'].str.replace('\\""','').str.replace('\\','')


data1['LastName'] = data1['LastName'].str.replace('\\""','').str.replace('\\','')


data1 = data1[['FirstName','LastName','Age','Email','Gender',"City",'Country']]
```

## **Data2**

```python

data2[['FirstName','LastName']] = data2['FirstName LastName'].str.split(expand=True)

data2.drop(['FirstName LastName'],axis=1, inplace=True)

data2['Country'] = "USA"

```

```python
data2["LastName"] = data2["LastName"].str.replace('"','').str.replace('\\','')

data2["FirstName"] = data2["FirstName"].str.replace('"','').str.replace('\\','')

data2['Age'] = data2['Age'].str.extract("(\d+)")

data2 = data2.astype({"Age": "int64"})

gender = {"0": "Female", "1" : "Male", "F" : "Female", "M" : "Male"}

data2['Gender'] = data2['Gender'].replace(gender)
```

```python
data2['City'] = data2['City'].str.capitalize()

data2['Email'] = data2['Email'].str.lower()

data2['FirstName'] = data2['FirstName'].str.capitalize()

data2['LastName'] = data2['LastName'].str.capitalize()


data2 = data2[['FirstName','LastName','Age','Email','Gender','City',"Country"]]
```

#  **Data3**






```python
data3= data3.replace(to_replace=r'string_|integer_|boolean_|character_', value='', regex=True)

data3['Age'] = data3['Age'].replace(to_replace=r'[a-zA-Z]', value='', regex=True)

data3["Country"] = "USA"
```

```python
data3[["FirstName", "LastName"]] = data3['Name'].str.split(expand=True)

data3.drop(["Name"], axis=1, inplace=True)
```

```python
data3['Email'] = data3['Email'].str.lower()

data3['Age'] = data3['Age'].str.replace('"', '')

data3["LastName"] = data3["LastName"].str.replace('"','')

data3['City'] = data3['City'].str.capitalize()

data3['LastName'] = data3['LastName'].str.capitalize()

data3['FirstName'] = data3['FirstName'].str.capitalize()



```

```python
gender = {"0": "Female", "1" : "Male", "F" : "Female", "M" : "Male"}

data3['Gender'] = data3['Gender'].replace(gender)

data3 = data3[['FirstName','LastName','Age','Email','Gender','City','Country']]

```

```python
dfull = [data1,data2,data3]

DataFull = pd.concat(dfull, ignore_index=True)

DataFull['City'] = DataFull['City'].str.replace('_',' ').str.replace('-',' ')

DataFull['Email'] = DataFull['Email'].str.replace('_',' ').str.replace('.',' ')
```
